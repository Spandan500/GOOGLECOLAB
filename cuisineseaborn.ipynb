**Line plot**:

*POV*:

It is useful for analysis of location vs food rating:

1. Easy comparison between locations-
Multiple lines can be drawn on the same graph, making it simple to compare food ratings of different locations.

  2.Helps identify increases and decreases-
Line plots clearly highlight rises and drops in food ratings, which helps in understanding impact on food quality or customer preference.

```python
sns.lineplot(x=dataset["Location"], y=dataset["Food Rating"])
plt.title("Lineplot Representation Of Data", fontsize=13, loc="left")
plt.xlabel("Location")
plt.ylabel("rating")
plt.xticks(rotation=45)
plt.grid(axis="both")
plt.show()
```

**BAR Plot**:

*POV*:
How it works-
1. Group Data: For each location, calculate the average overall food rating for each gender category.

2. Plot Bars: Draw a bar for each location.

3. Group by Gender: Within each location's bar, segments (or separate bars) are created, colored by gender to show their respective average ratings.

```python
plt.figure(figsize=(17, 7))
figure,axis = plt.subplots()
visual= sns.barplot(x= dataset['Location'], y=dataset['Overall Rating'], data= dataset, hue= 'Gender', palette='dark', ax=axis)
axis.bar_label(axis.containers[-1], fmt="%.1f", label_type="center")
axis.bar_label(axis.containers[-2], fmt="%.1f", label_type="center")
plt.title('Average Overall rating by Location')
plt.xlabel('Location')
plt.ylabel('Overall Rating')
plt.xticks(rotation = 90)
#plt.grid(axis = 'both')
plt.show()
```

**Count plot**:

*POV*:

To understand demographic preferences (Gender) within Location regarding cuisine types.
1. x='Location': Each bar group represents a specific area or city, showing the count of observations (e.g., restaurant visits, orders).

2. hue='Gender': Within each location's bars, different colors (hues) distinguish between 'Male' and 'Female' (or other categories), showing their relative frequencies.
3. The height of each colored segment shows the number of individuals of that gender in that location.

```python
plt.figure(figsize=(12,5))
visual= sns.countplot(x=dataset['Location'], hue="Gender", data=dataset, palette='dark')
plt.title("People count by location", fontsize=11, loc="left")
plt.xlabel("Location details")
plt.ylabel("People count")
plt.xticks(rotation = 60)
#annotate
for i in visual.patches:
  # print((i.get_x() + i.get_width()/2., i.get_height()))
  visual.annotate(
      format(i.get_height(), '.1f'),
      (i.get_x() + i.get_width()/2., i.get_height()),
      ha="center", #va="center",
      xytext=(3.5, 3.5),
      textcoords="offset points",
      fontsize=13, fontweight="normal"
  )
plt.show()
```

**Box Plot**:

*POV*:
1. x='Location': Each bar group represents a specific location.

**Scatterplot Of MATPLOTLIB:**

**Violinplot:**

*POV:*

The main goal is to compare the entire distribution of food ratings for each location.
1. X-axis: The categorical variable, representing different locations.
2. Y-axis: The numerical variable, representing the food rating scale (e.g., 1 to 5 stars).
3. The "Violin" Shape: The width of each "violin" at a specific rating level indicates the density or frequency of observations (restaurants) at that rating.
  a) Wider sections: Mean more restaurants in that location received that specific rating.
  b)Narrower sections: Mean fewer restaurants received that rating.
  c)Multiple peaks (bimodal distribution): The plot can reveal if a location has two distinct concentrations of ratings (e.g., many highly-rated and many low-rated places, but few average ones).

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="Location", y="Food Rating", hue="Gender", data=dataset, palette="dark")
plt.title(" Location vs Food rating", fontsize=13, loc="left")
plt.xlabel("locations")
plt.ylabel("food rating")
plt.xticks(rotation=45)
plt.show()
```

**Box plot**:

*POV*:

1.Here, each plot clearly shows the outliers and spread i.e. analysis of traits vs rating is plotted distinguishably.

```python
# List of categorical traits
traits = ['Gender', 'Marital Status', 'Activity','Smoker']
rating_col = 'Overall Rating'

# Set up the figure
plt.figure(figsize=(12, 9))

# Create a boxplot for each trait
for i, trait in enumerate(traits):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data=dataset, x=trait, y=rating_col, palette='dark')
    plt.title(f'{rating_col} by {trait}')
    plt.xlabel(trait)
    plt.ylabel(rating_col)
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

Key Aspects & Purpose: To quickly understand relationships and distributions across multiple features in your data like here "Gender" is the pivotal feature.

**BAR Plot:**

**CORRELATION ANALYSIS**
*Correlation: helps us to get relational scores upon every numerical column*

**Correlation Matrix Reading**

1.If Value (0.5 to 1) then Positive Correlation

2.If Value (0.0 to 0.5) then No Correlation

3.If Value (Less Than 0.0 / -0.35454) then negetive correlation



```python
plt.figure(figsize=(8, 6))
sns.heatmap(dataset[['Budget', 'Food Rating', 'Service Rating', 'Overall Rating', 'YOB']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
